**Implemention Of model**

```
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```
# Preprocessing Input data
data =pd.read_csv('data.csv')
GX = data.iloc[:,0]
GY = data.iloc[:,1]
plt.scatter(GX,GY)
plt.show()
```

```
# Building the model
m=0
c=0
l=0.0001
epochs = 1000
n = float(len(GX))
for i in range(epochs):
  y_pred = m *GX + c
  dm = (-2/n)*(sum(GX * (GY-y_pred)))
  dc = (-2/n)*(sum(GY-y_pred))
  m = m -l *dm
  c= c-l   *dc
print(m,c)
```

```
# Making predictions
y_pred = m *GX + c
plt.plot([min(GX), max(GX)], [min(y_pred), max(y_pred)], color='red')
plt.scatter(GX,GY)
plt.show()
```
